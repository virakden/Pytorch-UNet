```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


def compare_and_save_images(img1_path, img2_path, save_path):
    # Load the images
    img1 = cv2.imread(img1_path)
    img2 = cv2.imread(img2_path)

    # Convert images to RGB for displaying (since OpenCV loads images in BGR)
    img1_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
    img2_rgb = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

    # Create a figure for displaying the images side by side
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))

    # Plot Image 1 (left side)
    axes[0].imshow(img1_rgb)
    axes[0].set_title("Input Image2")
    axes[0].set_xlabel("X Coordinate")
    axes[0].set_ylabel("Y Coordinate")
    axes[0].axis("on")  # Keep axis on for X, Y labeling

    # Plot Image 2 (right side)
    axes[1].imshow(img2_rgb)
    axes[1].set_title("Output Mask2")
    axes[1].set_xlabel("X Coordinate")
    axes[1].set_ylabel("Y Coordinate")
    axes[1].axis("on")  # Keep axis on for X, Y labeling

    # Save the comparison image to a file
    plt.tight_layout()
    plt.savefig(save_path)

    # Display the images with labeled axes
    plt.show()

    print(f"Comparison image saved to: {save_path}")


# Example usage
compare_and_save_images(
    "data/imgs/0cdf5b5d0ce1_07.jpg",
    "data/predictions/image1_out.png",
    "data/compare/Image1.png",
)
```
